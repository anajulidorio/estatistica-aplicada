```python
def amostragem_sistematica(dataset, amostras): #crio uma função que posso utilizar a qualquer momento pra me retornar amostragens.
  intervalo=len(dataset)//amostras  #vai descobrir qual o intervalo ideal, dividindo a populacao pra saber a cada quantos vou retirar 1 elemento
  random.seed(1) #fixo seed para conseguir repetir sempre igual (dentro da mesma populacao)
  inicio=random.randint(0, intervalo) #pego aleatoriamento 1 item dentro do primeiro intervalo
  indices = np.arange(inicio, len(dataset), intervalo) #pego o restante dos elementos, indo até o final da populacao, pulando de acordo com intervalo encontrado
  amostra_sistematica=dataset.iloc[indices] #coloco o vetor feito dentro da variável
  return amostra_sistematica #retorno esse vetor para quem estiver chamando a função
```

```python
df_amostra_sistematica=amostragem_sistematica(dataset, 100) #chama a funcao que criamos, passando que quer 100 elementos
df_amostra_sistematica.shape #mostra que realmente retorno 100 elementos
```

```python
def amostragem_agrupamento(dataset, numero_grupos):  #estou criando uma função para fazer a divisão em grupos de forma automatizada
  intervalo=len(dataset)/numero_grupos

  grupos = []                  #crio um array pra colocar os grupos
  id_grupo = 0                 #Começo com o primeiro id sendo 0
  contagem = 0                 #Começo com o contador igual a 0
  for _ in dataset.iterrows(): # Começo meu loop para quebrar os itens em grupos, vai fazer do primeiro até o último item
    grupos.append(id_grupo)    #sempre que começar no loop, vai colocar mais um item com o valor do id_grupo(que começa com 0) no array
    contagem += 1              #Contagem vai aumentar de 1 em 1
    if contagem > intervalo:    #Quanto chegar passar o valor do intervalor, vai resetar a contagem pra zero, e aumentar em 1 o id_do grupo
      contagem = 0
      id_grupo += 1
  dataset['grupos'] = grupos    #criando uma nova coluna no dataset e colocando nela os valores do vetor que criamos.
  grupo_selecionado=random.randint(0,numero_grupos-1) #selecionando aleatoriamente um grupo
  return dataset[dataset['grupos']==grupo_selecionado] #retornando o grupo selecionado
```

```python
df_amostra_agrupamento = amostragem_agrupamento(dataset,100)
df_amostra_agrupamento.shape, df_amostra_agrupamento['grupos'].value_counts() #vai mostrar qual grupo foi selecionado e a quantidade de itens
```
